```python
# 중복 제거된 컬럼 리스트
columns_to_select = ['subject_id', 'deathtime', 'dod', 'charttime', 'eGFR_2021', 'SPARK_score','SPARK_class',
                     'hospital_stay_duration', 'ed_stay_duration', 'age', 'gender', 'serumCreatinine']

# 새로운 데이터프레임 생성
new_df = df[columns_to_select]

# 새로운 데이터프레임 확인
print(new_df.head())
```

```python
# Step 1: deathtime과 dod를 숫자형(int)으로 변환
new_df['deathtime'] = pd.to_numeric(new_df['deathtime'], errors='coerce').fillna(0).astype(int)
new_df['dod'] = pd.to_numeric(new_df['dod'], errors='coerce').fillna(0).astype(int)

# Step 2: deathtime과 dod 열의 or 연산을 통해 하나라도 1이면 1로 처리
new_df['deathtime'] = new_df['deathtime'] | new_df['dod']

# Step 3: dod 열 제거
new_df = new_df.drop(columns=['dod'])

# 결과 확인
print(new_df.head())

```

```python
# Step 1: SPARK_class를 기반으로 AKI 단계를 설정
def determine_aki_stage(row):
    if row['deathtime'] == 1:
        return 3  # Critical AKI
    elif row['SPARK_class'] in ['A', 'B']:  # SPARK_class 1, 2
        return 1  # No AKI
    elif row['SPARK_class'] in ['C', 'D']:  # SPARK_class 3, 4
        return 2  # Low-stage AKI
    else:
        return 0  # 예외 처리

# AKI_stage 설정
new_df['AKI_stage'] = new_df.apply(determine_aki_stage, axis=1)

# Step 2: SPARK_class와 deathtime은 삭제하지 않고, 독립변수로 사용하지 않을 예정이므로 저장만 유지
# 독립변수로 사용하지 않을 피처를 제거할 수 있도록 리스트에 따로 저장
unused_features = ['SPARK_class', 'deathtime']

# 결과 확인
print(new_df[['SPARK_class', 'deathtime', 'AKI_stage']].tail())
```

```python
import pandas as pd
import numpy as np

# 'charttime'을 datetime 형식으로 변환
new_df['charttime'] = pd.to_datetime(new_df['charttime'])

# 'subject_id'와 'charttime' 기준으로 정렬
new_df = new_df.sort_values(by=['subject_id', 'charttime'])

# 각 'subject_id'별로 시간 차이를 계산
new_df['time_diff'] = new_df.groupby('subject_id')['charttime'].diff()

# 2일 이상의 차이가 나는 경우 새로운 그룹으로 취급
new_df['new_group'] = (new_df['time_diff'] > pd.Timedelta(days=2)).cumsum()

# 새로운 그룹을 각 환자의 'subject_id'와 'new_group'으로 그룹화, subject_id 업데이트
new_df['group_id'] = new_df.groupby('subject_id')['new_group'].cumsum()
new_df['subject_id'] = new_df['subject_id'].astype(str) + '-' + new_df['group_id'].astype(str)

# 패딩 및 버려진 데이터 개수를 저장할 리스트
padding_count_list = []
discarded_count_list = []

# 각 'subject_id'마다 10시간 간격으로 데이터를 생성하고, 50개의 데이터를 만들기 위한 준비
def create_evenly_spaced_data(group):
    # 10시간 간격으로 시간 생성
    time_range = pd.date_range(start=group['charttime'].min(), end=group['charttime'].max(), freq='10H')

    # 시간 축으로 새로운 데이터프레임 생성
    time_df = pd.DataFrame({'charttime': time_range})

    # 보간할 수 없는 Categorical 열들을 따로 분리하고, charttime을 포함
    categorical_cols = group.select_dtypes(include=['category', 'object']).columns
    categorical_data = group[categorical_cols].copy()
    categorical_data['charttime'] = group['charttime']  # charttime을 포함하여 병합에 사용

    # 보간할 수 있는 열들만 따로 분리
    numeric_data = group.drop(columns=categorical_cols)

    # 시간 축과 병합하여 숫자형 데이터에 대한 등간격 데이터 생성 및 보간
    merged_df = pd.merge_asof(time_df, numeric_data.sort_values('charttime'), on='charttime')
    merged_df = merged_df.interpolate(method='linear')

    # Categorical 열을 가장 가까운 값으로 채움 (앞 방향으로 채움)
    categorical_merged_df = pd.merge_asof(time_df, categorical_data.sort_values('charttime'), on='charttime', direction='nearest')
    merged_df[categorical_cols] = categorical_merged_df[categorical_cols].ffill().bfill()

    # 패딩 및 버려진 데이터 수 계산
    padding_count = 0
    discarded_count = 0

    # 50개의 데이터가 부족하면 패딩(뒤쪽에 같은 행을 복사하여 추가)
    if len(merged_df) < 50:
        padding_rows = 50 - len(merged_df)
        padding_df = pd.DataFrame([merged_df.iloc[-1]] * padding_rows)  # 마지막 행 복사
        merged_df = pd.concat([merged_df, padding_df], ignore_index=True)
        padding_count = padding_rows
    elif len(merged_df) > 50:
        discarded_count = len(merged_df) - 50

    # 50개의 데이터로 맞춤
    merged_df = merged_df.iloc[:50]

    # 패딩 및 버린 데이터 수를 기록
    padding_count_list.append(padding_count)
    discarded_count_list.append(discarded_count)

    return merged_df

# 각 subject_id에 대해 위의 함수 적용
new_df_interpolated = new_df.groupby('subject_id').apply(lambda group: create_evenly_spaced_data(group.reset_index(drop=True))).reset_index(drop=True)

# 패딩 및 버려진 데이터에 대한 로그 출력
average_padding = np.mean(padding_count_list)
total_discarded = np.sum(discarded_count_list)

print(f"평균적으로 추가된 패딩 행 개수: {average_padding}")
print(f"총 버려진 데이터 개수: {total_discarded}")

# 결과 확인
print(new_df_interpolated.head())
```
